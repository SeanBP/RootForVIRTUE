```python
#!/usr/bin/env python3

"""
ePIC ROOT → VIRTUE Event Converter

Purpose
-------
Convert reconstructed ePIC simulation events stored in an EDM4eic ROOT file
into the JSON format used by the VIRTUE event display.

This example was developed for:
    e + p DIS simulations
    ePIC detector
    EIC software stack (eicrecon)

The script reads:
    • calorimeter hits
    • reconstructed clusters
    • reconstructed tracks
    • reconstructed jets
    • event kinematics

and writes:
    • VIRTUE-compatible JSON event files

This file is intended to serve as a template for future experiments.
Most detector-specific customization is isolated in the configuration
section and branch-selection logic.
"""

# ============================================================================
# Imports
# ============================================================================

import json
import math

import numpy as np
import uproot as ur


# ============================================================================
# User Configuration
# ============================================================================

EVENT_TYPE = "CCDIS"      # "NCDIS" or "CCDIS"
MIN_Q2 = 1000            # 100 or 1000

MAX_EVENTS = 20

INPUT_FILE = (
    f"RootFiles/"
    f"pythia8{EVENT_TYPE}_18x275_"
    f"minQ2={MIN_Q2}_"
    f"beamEffects_xAngle=-0.025_"
    f"hiDiv_1.0000.eicrecon.tree.edm4eic.root"
)

OUTPUT_FILE = (
    f"{EVENT_TYPE}_Q2={MIN_Q2}_Pythia8_{MAX_EVENTS}.json"
)

# Speed of light in mm/ns
C_LIGHT = 0.299792


# ============================================================================
# Utility Functions
# ============================================================================

def has_negative_value(array):
    """
    Return True if any nested value is negative.

    Several detector systems contain invalid timing information.
    Negative times are treated as missing values.
    """
    for subarray in array:
        for value in subarray:
            if value < 0:
                return True
    return False


def flatten_detector_collections(collections):
    """
    Convert

        detector → event → objects

    into

        event → objects

    Example:
        [BEMC_hits, EEMC_hits, HCAL_hits]

    becomes

        event_0_all_hits
        event_1_all_hits
        ...
    """
    return [
        [obj for detector in collections for obj in detector[event]]
        for event in range(len(collections[0]))
    ]


def filter_forward_jets(px, py, pz, energy):
    """
    Keep only jets with positive pz.

    In ePIC DIS studies, positive pz corresponds to the hadron-going
    (forward) direction.
    """

    filtered_px = []
    filtered_py = []
    filtered_pz = []
    filtered_e = []

    for i in range(len(pz)):

        mask = pz[i] > 0

        filtered_px.append(px[i][mask])
        filtered_py.append(py[i][mask])
        filtered_pz.append(pz[i][mask])
        filtered_e.append(energy[i][mask])

    return (
        np.array(filtered_px, dtype=object),
        np.array(filtered_py, dtype=object),
        np.array(filtered_pz, dtype=object),
        np.array(filtered_e, dtype=object),
    )


def calculate_angles(px, py, pz):
    """
    Convert momentum vector into spherical coordinates.
    """

    magnitude = np.sqrt(px**2 + py**2 + pz**2)

    theta = np.arccos(pz / magnitude)
    phi = np.arctan2(py, px)

    return theta, phi


def pseudorapidity_to_polar_ratio(theta):
    """
    dη/dθ conversion factor.

    Used to translate jet cone size into a VIRTUE opening angle.
    """

    return abs(
        -0.5
        * (1 / math.sin(theta / 2))
        * (1 / math.cos(theta / 2))
    )


def estimate_time_from_position(x, y, z):
    """
    Estimate hit arrival time using straight-line propagation.
    """

    distance = np.sqrt(
        (x / 1000.0) ** 2 +
        (y / 1000.0) ** 2 +
        (z / 1000.0) ** 2
    )

    return distance / C_LIGHT


# ============================================================================
# ROOT File Loading
# ============================================================================

print("Opening ROOT file...")
events = ur.open(f"{INPUT_FILE}:events")

print("Scanning branches...")

hit_branches = [
    key
    for key in events.keys()
    if key.endswith("RecHits.position.x")
]

cluster_branches = [
    key
    for key in events.keys()
    if key.endswith("Clusters.position.x")
    and "TruthClusters" not in key
]


# ============================================================================
# Jet Information
# ============================================================================

print("Loading reconstructed jets...")

jet_px = np.array(
    events["ReconstructedJets/ReconstructedJets.momentum.x"]
)
jet_py = np.array(
    events["ReconstructedJets/ReconstructedJets.momentum.y"]
)
jet_pz = np.array(
    events["ReconstructedJets/ReconstructedJets.momentum.z"]
)
jet_e = np.array(
    events["ReconstructedJets/ReconstructedJets.energy"]
)

jet_px, jet_py, jet_pz, jet_e = filter_forward_jets(
    jet_px,
    jet_py,
    jet_pz,
    jet_e,
)


# ============================================================================
# Track Information
# ============================================================================

print("Loading tracks...")

vx = np.array(
    events["CentralTrackVertices/CentralTrackVertices.position.x"]
)
vy = np.array(
    events["CentralTrackVertices/CentralTrackVertices.position.y"]
)
vz = np.array(
    events["CentralTrackVertices/CentralTrackVertices.position.z"]
)

q_over_p = np.array(
    events["CentralCKFTrackParameters/CentralCKFTrackParameters.qOverP"]
)

track_theta = np.array(
    events["CentralCKFTrackParameters/CentralCKFTrackParameters.theta"]
)

track_phi = np.array(
    events["CentralCKFTrackParameters/CentralCKFTrackParameters.phi"]
)


# ============================================================================
# Hit Extraction
# ============================================================================

print("Loading calorimeter hits...")

hit_x = []
hit_y = []
hit_z = []
hit_t = []
hit_e = []

lfh_x = []
lfh_y = []
lfh_z = []
lfh_t = []
lfh_e = []

for branch in hit_branches:

    is_lfhcal = "LFHCAL" in branch

    target_x = lfh_x if is_lfhcal else hit_x
    target_y = lfh_y if is_lfhcal else hit_y
    target_z = lfh_z if is_lfhcal else hit_z
    target_t = lfh_t if is_lfhcal else hit_t
    target_e = lfh_e if is_lfhcal else hit_e

    target_x.append(np.array(events[branch]))
    target_y.append(
        np.array(events[branch.replace(".position.x", ".position.y")])
    )
    target_z.append(
        np.array(events[branch.replace(".position.x", ".position.z")])
    )

    time_branch = branch.replace(".position.x", ".time")

    if (
        time_branch not in events
        or has_negative_value(np.array(events.get(time_branch, [])))
    ):
        target_t.append(
            [
                np.zeros_like(inner)
                for inner in np.array(events.get(branch, []))
            ]
        )
    else:
        target_t.append(np.array(events[time_branch]))

    energy_branch = branch.replace(".position.x", ".energy")

    if energy_branch in events:
        target_e.append(np.array(events[energy_branch]))

    elif "DRICH" in branch:
        target_e.append(
            [-2 * np.ones_like(inner) for inner in np.array(events[branch])]
        )

    else:
        target_e.append(
            [-1 * np.ones_like(inner) for inner in np.array(events[branch])]
        )

hit_x = flatten_detector_collections(hit_x)
hit_y = flatten_detector_collections(hit_y)
hit_z = flatten_detector_collections(hit_z)
hit_t = flatten_detector_collections(hit_t)
hit_e = flatten_detector_collections(hit_e)

lfh_x = flatten_detector_collections(lfh_x)
lfh_y = flatten_detector_collections(lfh_y)
lfh_z = flatten_detector_collections(lfh_z)
lfh_t = flatten_detector_collections(lfh_t)
lfh_e = flatten_detector_collections(lfh_e)


# ============================================================================
# Cluster Extraction
# ============================================================================

print("Loading clusters...")

cluster_x = []
cluster_y = []
cluster_z = []
cluster_t = []
cluster_e = []
cluster_is_hcal = []

for branch in cluster_branches:

    cluster_x.append(np.array(events[branch]))
    cluster_y.append(
        np.array(events[branch.replace(".position.x", ".position.y")])
    )
    cluster_z.append(
        np.array(events[branch.replace(".position.x", ".position.z")])
    )

    time_branch = branch.replace(".position.x", ".time")

    if (
        time_branch not in events
        or has_negative_value(np.array(events.get(time_branch, [])))
    ):
        cluster_t.append(
            [
                np.zeros_like(inner)
                for inner in np.array(events.get(branch, []))
            ]
        )
    else:
        cluster_t.append(np.array(events[time_branch]))

    is_hcal = "Hcal" in branch or "HCAL" in branch

    cluster_is_hcal.append(
        [
            np.full_like(inner, is_hcal, dtype=bool)
            for inner in np.array(events[branch])
        ]
    )

    cluster_e.append(
        np.array(events[branch.replace(".position.x", ".energy")])
    )

cluster_x = flatten_detector_collections(cluster_x)
cluster_y = flatten_detector_collections(cluster_y)
cluster_z = flatten_detector_collections(cluster_z)
cluster_t = flatten_detector_collections(cluster_t)
cluster_e = flatten_detector_collections(cluster_e)
cluster_is_hcal = flatten_detector_collections(cluster_is_hcal)


# ============================================================================
# Event Kinematics
# ============================================================================

Q2 = np.array(events["InclusiveKinematicsTruth.Q2"])
bjorken_x = np.array(events["InclusiveKinematicsTruth.x"])


# ============================================================================
# VIRTUE Header
# ============================================================================

header = {
    "version": "3.0.0",
    "experiment": (
        f"pythia8 {EVENT_TYPE}_18x275 "
        f"minQ2 = {MIN_Q2} GeV2"
    ),
    "energy_unit": "GeV",
    "color_bar": "Log",
    "length_unit": "mm",
    "particles": [
        {
            "angle_rad": [0.0, 0.0],
            "size": 150.0,
            "color_rgba": [1.0, 0.0, 0.0, 1.0],
        },
        {
            "angle_rad": [3.14159, 0.0],
            "size": 100.0,
            "color_rgba": [0.0, 0.0, 1.0, 1.0],
        },
    ],
    "tracker_settings": {
        "B_field_T": 1.7,
        "tracker_boundary": [
            0.7817 * 1000,
            -1.235 * 1000,
            1.88 * 1000,
        ],
    },
}

output = {
    "header": header,
    "events": [],
}

# ============================================================================
# Event Construction
# ============================================================================

print("Building events...")

# ============================================================================
# Construct VIRTUE Events
# ============================================================================
#
# At this point all detector information has been loaded from the ROOT file
# and reorganized into event-by-event arrays.
#
# The goal of this section is to convert those detector objects into the
# VIRTUE JSON schema:
#
#     ROOT EDM4eic Objects
#            ↓
#     Python Event Objects
#            ↓
#     VIRTUE JSON
#
# Each event may contain:
#
#     • calorimeter hits
#     • reconstructed tracks
#     • reconstructed clusters
#     • reconstructed jets
#     • LFHCAL blocks
#
# The final result is stored in:
#
#     data["events"]
#
# ============================================================================



# Determine how many events to export.
#
# VIRTUE files can become quite large, so the original notebook limits
# the export to the first 20 events. If fewer than 20 events exist in
# the ROOT file, export all available events.

num_events = min(MAX_EVENTS, len(hit_x))
events = []
# --------------------------------------------------------------------------
# Loop over events
# --------------------------------------------------------------------------

for event_index in range(num_events):

    # ======================================================================
    # Determine Event Energy Scale
    # ======================================================================
    #
    # VIRTUE can use a per-event energy scale for color mapping.
    #
    # Collect all valid calorimeter energies from:
    #
    #     • standard calorimeter hits
    #     • LFHCAL blocks
    #
    # and determine the minimum and maximum energies present.
    #

    valid_energies = [
        energy
        for energy in e[i] + LFHe[i]
        if energy > 0
    ]

    # Default fallback values in case an event contains no energy deposits.

    min_energy = 1.0
    max_energy = 100.0

    if valid_energies:
        min_energy = float(min(valid_energies))
        max_energy = float(max(valid_energies))

    # ======================================================================
    # Create Event Metadata
    # ======================================================================

    info_text = f"Event #{i}"

    # Neutral-current DIS samples contain truth-level kinematic information
    # that can be displayed directly in the event viewer.

    if event_type == "NCDIS":

        info_text += (
            f", Q2 = {round(float(Q2[i][0]), 3)} GeV2"
            f", x = {round(float(bx[i][0]), 3)}"
        )

    event = {
        "event_data": {
            "info_text": info_text,
            "energy_scale": [min_energy, max_energy]
        },
        "hits": [],
        "clusters": [],
        "tracks": [],
        "jets": [],
        "blocks": []
    }

    # ======================================================================
    # Build Hit Objects
    # ======================================================================
    #
    # Energy is converted into a logarithmic scale between 0 and 1:
    #
    #     normalized = log(E / Emin) / log(Emax / Emin)
    #
    # This improves visibility when the event contains deposits spanning
    # many orders of magnitude.
    #
    # Special energy codes:
    #
    #     -1 : detector hit without energy information
    #     -2 : DRICH photon hit
    #

    normalized_energy = [
        (
            np.log(energy / min_energy)
            / np.log(max_energy / min_energy)
        )
        if energy > 0 else energy
        for energy in e[i]
    ]

    hits = []

    for j in range(len(x[i])):

        energy = float(normalized_energy[j])

        # ------------------------------------------------------------------
        # Assign hit color
        # ------------------------------------------------------------------
        #
        # Normal hits:
        #
        #     low energy  → blue
        #     high energy → red
        #
        # Missing-energy hits:
        #
        #     black
        #
        # DRICH hits:
        #
        #     white
        #

        if energy == -1:

            red, green, blue, alpha = 0, 0, 0, 1

        elif energy == -2:

            red, green, blue, alpha = 1, 1, 1, 1

        else:

            red = energy
            green = 0
            blue = 1 - energy
            alpha = energy

        # Hits without energy information are drawn smaller.

        size = 50 if energy == -1 else 100

        # ------------------------------------------------------------------
        # Determine hit time
        # ------------------------------------------------------------------
        #
        # Some detectors do not provide valid timing information.
        #
        # If time < 0.1 ns, estimate the arrival time assuming travel at
        # the speed of light from the interaction point.
        #

        if t[i][j] < 0.1:

            dist = np.sqrt(
                (x[i][j] / 1000.0) ** 2 +
                (y[i][j] / 1000.0) ** 2 +
                (z[i][j] / 1000.0) ** 2
            )

            time = dist / 0.299792

        else:

            time = t[i][j]

        hit = {
            "position": [
                float(x[i][j]),
                float(y[i][j]),
                float(z[i][j])
            ],
            "time_ns": float(time),
            "size": float(size),
            "color_rgba": [
                float(red),
                float(green),
                float(blue),
                float(alpha)
            ]
        }

        hits.append(hit)

    event["hits"] = hits

    # ======================================================================
    # Build Track Objects
    # ======================================================================
    #
    # Tracks are represented by:
    #
    #     qOverP
    #     theta
    #     phi
    #     production vertex
    #
    # VIRTUE reconstructs the helical trajectory internally using the
    # detector magnetic field specified in the file header.
    #

    tracks = []

    if len(vx[i]) > 0:

        evx = float(vx[i][0])
        evy = float(vy[i][0])
        evz = float(vz[i][0])

    else:

        evx = 0.0
        evy = 0.0
        evz = 0.0

    for j in range(len(qOverP[i])):

        track = {
            "qOverP": float(qOverP[i][j]),
            "angle_rad": [
                float(Ttheta[i][j]),
                float(Tphi[i][j])
            ],
            "vertex": [evx, evy, evz],
            "duration_ns": [0.0, 1000.0],
            "color_rgba": [0.0, 1.0, 1.0, 1.0]
        }

        tracks.append(track)

    event["tracks"] = tracks

    # ======================================================================
    # Build Cluster Objects
    # ======================================================================
    #
    # Cluster length in VIRTUE is proportional to reconstructed energy.
    #
    # HCAL and ECAL clusters are drawn using different colors:
    #
    #     HCAL → cyan
    #     ECAL → green
    #

    clusters = []

    for j in range(len(cx[i])):

        energy = float(100 * ce[i][j])

        is_hcal = cc[i][j]

        if is_hcal:

            red, green, blue, alpha = (0, 1, 1, 0.5)

        else:

            red, green, blue, alpha = (0, 1, 0, 0.5)

        cluster = {
            "position": [
                float(cx[i][j]),
                float(cy[i][j]),
                float(cz[i][j])
            ],
            "time_ns": float(ct[i][j]),
            "granularity": 100.0,
            "length": energy,
            "color_rgba": [
                float(red),
                float(green),
                float(blue),
                float(alpha)
            ]
        }

        # Small clusters are suppressed to reduce visual clutter.

        if energy > 100:

            clusters.append(cluster)

    event["clusters"] = clusters

    # ======================================================================
    # Build Jet Objects
    # ======================================================================
    #
    # Jets are displayed as cones.
    #
    # The jet direction is determined from the reconstructed momentum
    # vector and converted into spherical coordinates.
    #

    jets = []

    for j in range(len(jetPx[i])):

        energy = float(80 * jetE[i][j])

        theta, phi = calculate_angles(
            jetPx[i][j],
            jetPy[i][j],
            jetPz[i][j]
        )

        op_angle = (
            0.5 /
            pseudorapidity_to_polar_ratio(theta)
        )

        # Approximate animation timing based on jet energy.

        dist = float(60 * jetE[i][j] / 1000.0)
        time = dist / 0.299792

        jet = {
            "length": energy,
            "R_rad": float(op_angle),
            "angle_rad": [
                float(theta),
                float(phi)
            ],
            "color_rgba": [0.0, 0.3, 1.0, 0.5],
            "time_ns": float(time)
        }

        jets.append(jet)

    event["jets"] = jets

    # ======================================================================
    # Build LFHCAL Blocks
    # ======================================================================
    #
    # LFHCAL cells are drawn as rectangular blocks rather than point-like
    # hits. The dimensions correspond approximately to the calorimeter
    # segmentation used in the detector geometry.
    #

    LFHnormalized_energy = [
        (
            np.log(energy / min_energy)
            / np.log(max_energy / min_energy)
        )
        if energy > 0 else energy
        for energy in LFHe[i]
    ]

    blocks = []

    for j in range(len(LFHx[i])):

        energy = float(LFHnormalized_energy[j])

        red = energy
        green = 0
        blue = 1 - energy
        alpha = energy

        dist = np.sqrt(
            (LFHx[i][j] / 1000.0) ** 2 +
            (LFHy[i][j] / 1000.0) ** 2 +
            (LFHz[i][j] / 1000.0) ** 2
        )

        time = dist / 0.299792

        block = {
            "position": [
                float(LFHx[i][j]),
                float(LFHy[i][j]),
                float(LFHz[i][j])
            ],
            "time_ns": float(time),
            "size": [
                50.0,
                50.0,
                float(1400 / 7)
            ],
            "color_rgba": [
                float(red),
                float(green),
                float(blue),
                float(alpha)
            ]
        }

        blocks.append(block)

    event["blocks"] = blocks

    # ======================================================================
    # Store Event
    # ======================================================================

    events.append(event)

# Attach all events to the final JSON structure.

data["events"] = events


# ============================================================================
# Write JSON
# ============================================================================

print(f"Writing {OUTPUT_FILE}")

with open(OUTPUT_FILE, "w") as outfile:
    json.dump(output, outfile, indent=4)

print("Done.")
```
